```python
class Hashabledict(dict):
    def __hash__(self):
        return hash(frozenset(self))
```

```python
class Hashableset(set):
    def __hash__(self):
        return hash(frozenset(self))
```

```python
def fff(row):
    c = Counter()
    
    for n in range(0,11):
        position = row['lineup_%02.d_position' % (n+1)]
        c[position] += 1
    return Hashableset(c)
```

```python
dtt['formation'] = dtt.apply(fff, axis=1)
```
