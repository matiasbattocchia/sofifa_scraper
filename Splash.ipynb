```python
import scrapy
from scrapy.http import TextResponse

import scrapy_splash

import requests

SPLASH_URL = 'http://localhost:8050'
```

```python
script = """
splash:go(args.url)
local e = splash:select('td.val')
ok, reason = e:mouse_click()
return splash:html()
"""

url = 'https://chinese.yabla.com/chinese-pinyin-chart.php'

r = requests.post('http://localhost:8050/run', json={
    'lua_source': script,
    'url': url
})

response = TextResponse(r.url, body=r.text, encoding='utf-8')
```

```python
response.xpath('//td[@class="val"]/text()').extract()
```
